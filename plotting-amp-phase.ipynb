#### Amplitude and phase graph of the given frequency response was plotted.
#### H(e^jw) = 1 - (0.45e^-jw)/(1-0.8e^-jw) 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
pi = np.pi
exp = np.exp
```

```python
w = np.linspace(0, 2*pi)
h = 1 - (0.45*exp(-1j*w))/(1-0.8*exp(-1j*w))
h_amp = np.abs(h)
h_phase = np.angle(h)

#--------------------------------------------------------------#

plt.figure() # Ã§izilecek yeni grafik iÃ§in
plt.plot(w/pi,h_amp)
plt.title('amplitude')
plt.show()

plt.figure() # Ã§izilecek yeni grafik iÃ§in
plt.plot(w/pi,h_phase)
plt.title('phase')
plt.show()
```

---
#### Amplitude and phase graph of the given frequency response was plotted.
#### X(e^jw) = 1/(1-ae^-jw), a = -0.5

```python
x = 1/(1-(-0.5)*exp(-1j*w))
x_amp = np.abs(x)
x_phase = np.angle(x)

#--------------------------------------------------------------#

plt.figure()
plt.plot(w/pi, x_amp)
plt.title('amplitude')
plt.show()

plt.figure()
plt.plot(w/pi, x_phase)
plt.title('phase')
plt.show()
```
