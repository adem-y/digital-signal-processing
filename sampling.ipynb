# SORU 1

$x_{1}(t)=7cos(200\pi t)$

$x_{2}(t)=3cos(200\pi t) + 4cos(1800\pi t)$

##### Fs = 2200 Hz

# soru1 a)
Bu soruda x1(t), x2(t) işaretleri 2200 Hz ile örneklenerek x1[n] ve x2[n] işaretleri elde edildi. İki işaret de nyquist teoremine göre Fs>=2Fn ifadesini sağlamaktadır bu yüzden örtüşme olmamaktadır farklı işaretler elde edilmiştir.

```python
import matplotlib.pyplot as plt
import numpy as np
cos = np.cos
pi = np.pi
```

```python
#--------x1(t) sinyalinin çizdirilmesi--------#

F = 100 # x1(t)'nin temel frekansı 100 Hz
T = 1/F
t = np.arange(0., 2*T, 0.0001)
x1 = 7*cos(200*pi*t)

plt.xlabel("t")
plt.title("x1(t)=7*cos(200*pi*t)")
plt.plot(t,x1)
plt.show()

#--------x1[n] sinyalinin çizdirilmesi--------#

Fs = 2200 # örnekleme frekansı
Ts = 1/Fs # örnekleme periyodu
N = 22 # örnekleme sayısı, k = 1 için
n = np.arange(0., 2*N) # örnekleme indisi, 2 periyot
x1n = 7*cos(200*pi*n*Ts) # örneklenmiş x1[n] işareti

plt.stem(n,x1n)
plt.title("x1[n]")
plt.xlabel("örnek")
plt.ylabel("x1[n]")
plt.show()
```

```python
#--------x2(t) sinyalinin çizdirilmesi--------#

F = 900 # x2(t)'nin temel frekansı 900 Hz
T = 1/F
t = np.arange(0., 2*T, 0.0001)
x2 = 3*cos(200*pi*t) + 4*cos(1800*pi*t)

plt.xlabel("t")
plt.title("x2(t)=3*cos(200*pi*t) + 4*cos(1800*pi*t)")
plt.plot(t,x2)
plt.show()

#--------x2[n] sinyalinin çizdirilmesi--------#

Fs = 2200 # örnekleme frekansı
Ts = 1/Fs # örnekleme periyodu
N = 22 # örnekleme sayısı
n = np.arange(0., 2*N) # örnekleme indisi, 2 periyot
x2n = 3*cos(200*pi*n*Ts) + 4*cos(1800*pi*n*Ts) # örneklenmiş x2[n] işareti

plt.stem(n,x2n)
plt.title("x2[n]")
plt.xlabel("örnek")
plt.ylabel("x2[n]")
plt.show()
```

# soru1 b)

```python
#-----x1[n]'in fourier transformu, x1(𝜔)-----#

w = np.linspace(-pi, pi-(2*pi/N), N) # -pi, pi aralığında N örnek
x1_w = np.fft.fft(x1n, N)/N # fourier transform
x1_w = np.fft.fftshift(x1_w) # x1(w) işaretinin fftshift ile -pi,+pi aralığında gösterilmesi

#----- x1(w) işaretinin grafiğinin çizdirilmesi -----#

for i in range(0, 22):
    if np.abs(x1_w)[i] == 3.5:
        print("w={}, genlik değeri={}".format(w[i]/pi, np.abs(x1_w)[i]))

plt.stem(w,np.abs(x1_w)) # x1(w)'nin genliğinin çizdirilmesi
plt.title("X1(w)'nın Genlik Grafiği")
plt.xlabel("w")
plt.show()

plt.stem(w,np.angle(x1_w)) # x1(w)'nin fazının çizdirilmesi
plt.title("X1(w)'nın Faz Grafiği")
plt.xlabel("w")
plt.show()
```

```python
#-----x2[n]'in fourier transformu, x2(𝜔)-----#

w = np.linspace(-pi, pi-(2*pi/N), N) # -pi, pi aralığında N örnek
x2_w = np.fft.fft(x2n, N)/N # fourier transform
x2_w = np.fft.fftshift(x2_w) # x2(w) işaretinin fftshift ile -pi,+pi aralığında gösterilmesi

#----- x2(w) işaretinin grafiğinin çizdirilmesi -----#

for i in range(0, 22):
    if np.abs(x2_w)[i] > 0.4:
        print("w = ", w[i]/pi)
        print("genlik = ", np.abs(x2_w)[i])
        print("faz = ", np.angle(x2_w)[i])
        print("---")

plt.figure()
plt.stem(w,np.abs(x2_w)) # x2(w)'nin genliğinin çizdirilmesi
plt.title("X2(w)'nın Genlik Grafiği")
plt.xlabel("w")
plt.show()

plt.figure()
plt.stem(w,np.angle(x2_w)) # x2(w)'nin fazının çizdirilmesi
plt.title("X2(w)'nın Faz Grafiği")
plt.xlabel("w")
plt.show()
```

# soru1 c)

$X1(\omega)$ işareti
##### w1 = -0.9$\pi$ 'de 3.5 genlik değerine,
##### w2 = 0.9$\pi$ 'de 3.5 genlik değeri gelmektedir.
##### Faz grafiğinde w1 ve w2 değerlerinin sıfır olduğu gözlemlenmiştir.
##### Bu durumda $x1[n] = 7$*$cos(0.09\pi n)$ gelmektedir.
##### Sinyal Fs = 2200 Hz ile örneklendiğinde $x1[n] = 7*cos(\pi n/11)$ gelmektedir. $0.09 = 1/11$ Aynı işaretin elde edilme sebebi örtüşme olmamasıdır. Nyquist teoremine göre: $Fs>2f_1$, $2200>=200$

$X2(\omega)$ işareti
##### w1 = -0.82$\pi$'de 2.0 genlik değerine,
##### w2 = 0.82$\pi$'de 2.0 genlik değerine,
##### w3 = -0.09$\pi$'de 1.5 genlik değeri,
##### w4 = 0.09$\pi$'de 1.5 genlik değeri gelmektedir.
##### Faz grafiğinde w1, w2, w3 ve w4 değerlerinin sıfır olduğu gözlemlenmiştir.
##### Bu durumda x2[n] = 4$*$cos(0.82n) + 3$*$cos(0.09n)
##### Sinyal Fs = 2200 Hz ile örneklendiğinde $x2[n] = 4*cos(9\pi n/11) + 3*cos(\pi n/11)$ gelmektedir. Nyquist teoremine göre örtüşme olmadığı için aynı sinyal elde edilmiştir.

# SORU 3
$x_{3}(t) = sin(9600\pi t)$ işareti Fs = 8192 Hz ile örneklenerek $x_{3}[n]$ işareti elde edilmektedir.

# soru3 a)
$x_{3}(t)$ işareti tam 3 periyot çizdirildi.

```python
import matplotlib.pyplot as plt
import numpy as np
sin = np.sin
pi = np.pi

F = 4800 # sinyalin temel frekansı
T = 1/F
# nyquist şartı sağlanmamaktadır. örtüşme meydana gelecektir.

t = np.arange(0., 3*T, 0.00001) # 3 tam periyot
x3 = sin(9600*pi*t)

plt.xlabel("t")
plt.title("x3(t)=sin(9600*pi*t)")
plt.plot(t,x3)
plt.show()
```

# soru3 b)
$x_{3}(t)$ işareti örneklenerek $x_{3}[n]$ oluşturuldu.

```python
Fs = 8192 # örnekleme frekansı
Ts = 1/Fs # örnekleme periyodu
N = 128 # örnekleme sayısı, k=75

n = np.arange(0., 3*N) # 3 periyot
x3n = sin(9600*pi*n*Ts) # örneklenmiş x3[n] işareti

plt.stem(n,x3n)
plt.title("x3[n]")
plt.xlabel("örnek")
plt.ylabel("x3[n]")
plt.show()
```

# soru3 c)
$x_{3}[n]$ işaretinin fft() fonksiyonu ile fourier dönüşümü elde edildi.

```python
w = np.linspace(-pi, pi-(2*pi/N), N) # -pi, pi aralığında N örnek
x3_w = np.fft.fft(x3n, N)/N # fourier transform
x3_w = np.fft.fftshift(x3_w) # x1(w) işaretinin fftshift ile -pi,+pi aralığında gösterilmesi

for i in range(0, 128):
    if np.abs(x3_w)[i] > 0.4:
        print("w = ", w[i]/pi)
        print("genlik = ", np.abs(x3_w)[i])
        print("faz = ", np.angle(x3_w)[i])
        print("---")

plt.figure()
plt.stem(w,np.abs(x3_w)) # x1(w)'nin genliğinin çizdirilmesi
plt.title("X3(w)'nın Genlik Grafiği")
plt.xlabel("w")
plt.show()

plt.figure()
plt.stem(w,np.angle(x3_w)) # x1(w)'nin fazının çizdirilmesi
plt.title("X3(w)'nın Faz Grafiği")
plt.xlabel("w")
plt.show()
```

# SORU 4
---
x[n] = [1.25, 3.5, 4, 6.7, -1.4, -3.5, -10.5, 8.5, 3.8]

```python
import numpy as np
import matplotlib.pyplot as plt

xn = np.array([1.25, 3.5, 4, 6.7, -1.4, -3.5, -10.5, 8.5, 3.8])
n = np.arange(0, len(xn))
plt.figure()
plt.stem(n, xn)
```

```python
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    
    return quantized.reshape(x.shape)
```

```python
k_bit = 3
Max_range = np.amax(xn) - np.amin(xn) # işaretin y ekseninde aldığı max değer aralığı
q = pow(2, k_bit) # seviye sayısı
delta = Max_range/q # adım aralığı

S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta # mid-rise yöntemi
S_midtread = -np.amax(xn) + np.arange(q)*delta # mid-tread yöntemi

#--- kuantalanmış işaretlerin elde edilmesi ---#
y_midtrise = quantize(xn, S_midrise)
y_midtread = quantize(xn, S_midtread)
```

```python
plt.figure()
fig, ax = plt.subplots()
ax.stem(n,xn)
ax.bar(n,y_midtrise,color='orange',edgecolor = "darkorange")
plt.xlabel("örnek")
plt.title('midrise yöntemiyle kuantalanmış işaret')

fig, ax = plt.subplots()
ax.stem(n,xn)
ax.bar(n,y_midtread,color='orange',edgecolor = "darkorange")
plt.xlabel("örnek")
plt.title('midtread yöntemiyle kuantalanmış işaret')
plt.show()
```
