#### ùë•[ùëõ]=‚Å°2sin‚Å°(2ùúãùëõ/7‚àíùúã/5) is plotted for 2 full periods.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

n = np.arange(0, 13, 1)
x = np.zeros(len(n))
pi = np.pi
sin = np.sin

for k in range(len(n)):
    x[k] = 2*sin(2*k*pi/7-pi/5)
    
plt.stem(n, x, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$ùë•[ùëõ] =‚Å° 2sin‚Å°(2ùúãùëõ/7‚àíùúã/5)$')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.show()
```

#### x2[n] = exp(1j*pi/5) is plotted in range [0,15]

```python
plt.figure()
exp = np.exp

n = np.arange(0, 16, 1)
x2 = np.zeros(len(n), dtype=complex)

for k in range(len(n)):
    x2[k] = exp(1j*pi/5)
    
plt.stem(n, x2.real, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x2[n] = exp(jùúã/5)$ - real part')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.show()

plt.figure() # yeni grafik i√ßin
plt.stem(n, x2.imag, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x2[n] = exp(jùúã/5)$ - imag part')
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.show()
```

#### x3[n] = exp(jùúãùëõ/5) is plotted in range [0,15]. It has been observed from the graph of this signal that it is a periodic signal. The signal oscillates due to the complex number of the expression.

```python
plt.figure()

n = np.arange(0, 16, 1)
x3 = np.zeros(len(n), dtype=complex)

for k in range(len(n)):
    x3[k] = exp(1j*pi*k/5)
    
plt.stem(n, x3.real, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x3[n] = exp(jùúãùëõ/5)$ - real kƒ±smƒ±')
plt.xlabel('n')
plt.ylabel('x3[n]')
plt.show()

plt.figure()
plt.stem(n, x3.imag, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x3[n] = exp(jùúãùëõ/5)$ - imag kƒ±smƒ±')
plt.xlabel('n')
plt.ylabel('x3[n]')
plt.show()
```

#### x4[n] = cos(ùúã/5) + jsin(ùúã/5) is plotted in range [0,15]. Euler's equation is the reason why this graph is exactly the same as x2[n]. e^ix = cosx + isinx

```python
plt.figure()

n = np.arange(0, 16, 1)
x4 = np.zeros(len(n), dtype=complex)

cos = np.cos

for k in range(len(n)):
    x4[k] = cos(pi/5) + 1j*sin(pi/5)
    
plt.stem(n, x4.real, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x4[n] = cos(ùúã/5) + jsin(ùúã/5)$ - real part')
plt.xlabel('n')
plt.ylabel('x4[n]')
plt.show()

plt.figure()
plt.stem(n, x4.imag, use_line_collection=True)
plt.xticks(np.arange(n[0], n[-1], 1))
plt.title('$x4[n] = cos(ùúã/5) + jsin(ùúã/5)$ - imag part')
plt.xlabel('n')
plt.ylabel('x4[n]')
plt.show()
```

#### x5[n] = cos(ùúãùëõ/5) + jsin(ùúãùëõ/5) is plotted in range [0,15]. Euler's equation is the reason why this graph is exactly the same as x3[n]. e^ix = cosx + isinx
